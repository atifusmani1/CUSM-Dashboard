```python
# Create reorganized dataframes with intuitive names and structure

# EBSCO Database Usage
ebsco_db_usage = df1.copy()
# Clean column names
ebsco_db_usage.columns = [col.strip().lower().replace(' ', '_') for col in ebsco_db_usage.columns]
# Sort by date if available, otherwise by database name
if 'reporting_period' in ebsco_db_usage.columns:
    ebsco_db_usage = ebsco_db_usage.sort_values('reporting_period')
elif 'database' in ebsco_db_usage.columns:
    ebsco_db_usage = ebsco_db_usage.sort_values('database')

# EBSCO Platform Usage
ebsco_platform_usage = df2.copy()
# Clean column names
ebsco_platform_usage.columns = [col.strip().lower().replace(' ', '_') for col in ebsco_platform_usage.columns]
# Sort by date if available
if 'reporting_period' in ebsco_platform_usage.columns:
    ebsco_platform_usage = ebsco_platform_usage.sort_values('reporting_period')

# EBSCO Title Usage
ebsco_title_usage = df3.copy()
# Clean column names
ebsco_title_usage.columns = [col.strip().lower().replace(' ', '_') for col in ebsco_title_usage.columns]
# Sort by title
if 'title' in ebsco_title_usage.columns:
    ebsco_title_usage = ebsco_title_usage.sort_values('title')

# Platform Master Report
platform_master = df4.copy()
# Clean column names
platform_master.columns = [col.strip().lower().replace(' ', '_') for col in platform_master.columns]
# Sort by platform name if available
if 'platform' in platform_master.columns:
    platform_master = platform_master.sort_values('platform')

# Sensource Commons Usage
commons_usage = df5.copy()
# Clean column names
commons_usage.columns = [col.strip().lower().replace(' ', '_') for col in commons_usage.columns]
# Sort by date if available
if 'date' in commons_usage.columns:
    commons_usage = commons_usage.sort_values('date')

# Combined EBSCO usage (if possible)
# Note: This is a placeholder - we'll need to see the actual data structure to properly join
try:
    # Attempt to create a combined usage dataframe if the data structures allow it
    # This will depend on having common columns to join on
    combined_ebsco = pd.DataFrame()
    
    # Check if we can join database and platform data
    common_cols_db_platform = set(ebsco_db_usage.columns).intersection(set(ebsco_platform_usage.columns))
    if len(common_cols_db_platform) > 0:
        join_col = list(common_cols_db_platform)[0]  # Use first common column as join key
        combined_ebsco = pd.merge(ebsco_db_usage, ebsco_platform_usage, on=join_col, suffixes=('_db', '_platform'))
    
    # Check if we can add title data
    if not combined_ebsco.empty:
        common_cols_combined_title = set(combined_ebsco.columns).intersection(set(ebsco_title_usage.columns))
        if len(common_cols_combined_title) > 0:
            join_col = list(common_cols_combined_title)[0]
            combined_ebsco = pd.merge(combined_ebsco, ebsco_title_usage, on=join_col, suffixes=('', '_title'))
except Exception as e:
    print(f"Could not create combined EBSCO dataframe: {e}")
    combined_ebsco = pd.DataFrame()  # Empty dataframe if combination fails
```

```python
# Create additional analysis-ready dataframes

# 1. Monthly usage summary (if date information is available)
try:
    # For EBSCO database usage
    if 'reporting_period' in ebsco_db_usage.columns and 'searches_regular' in ebsco_db_usage.columns:
        monthly_db_usage = ebsco_db_usage.groupby('reporting_period').agg({
            'searches_regular': 'sum',
            'total_item_requests': 'sum' if 'total_item_requests' in ebsco_db_usage.columns else None
        }).reset_index().sort_values('reporting_period')
        print("Monthly Database Usage Summary:")
        print(monthly_db_usage)
        print("\n" + "="*80 + "\n")
    
    # For platform usage
    if 'reporting_period' in ebsco_platform_usage.columns and 'searches_regular' in ebsco_platform_usage.columns:
        monthly_platform_usage = ebsco_platform_usage.groupby('reporting_period').agg({
            'searches_regular': 'sum',
            'total_item_requests': 'sum' if 'total_item_requests' in ebsco_platform_usage.columns else None
        }).reset_index().sort_values('reporting_period')
        print("Monthly Platform Usage Summary:")
        print(monthly_platform_usage)
        print("\n" + "="*80 + "\n")
        
    # For Commons usage
    if 'date' in commons_usage.columns:
        # Extract month from date if it's a datetime column
        if pd.api.types.is_datetime64_any_dtype(commons_usage['date']):
            commons_usage['month'] = commons_usage['date'].dt.to_period('M')
            monthly_commons_usage = commons_usage.groupby('month').agg({
                col: 'sum' for col in commons_usage.select_dtypes(include=['number']).columns
            }).reset_index().sort_values('month')
            print("Monthly Commons Usage Summary:")
            print(monthly_commons_usage)
            print("\n" + "="*80 + "\n")
except Exception as e:
    print(f"Could not create monthly summaries: {e}")

# 2. Top resources by usage (if applicable)
try:
    # Top databases by searches
    if 'database' in ebsco_db_usage.columns and 'searches_regular' in ebsco_db_usage.columns:
        top_databases = ebsco_db_usage.groupby('database').agg({
            'searches_regular': 'sum',
            'total_item_requests': 'sum' if 'total_item_requests' in ebsco_db_usage.columns else None
        }).reset_index().sort_values('searches_regular', ascending=False).head(10)
        print("Top 10 Databases by Usage:")
        print(top_databases)
        print("\n" + "="*80 + "\n")
    
    # Top titles by requests
    if 'title' in ebsco_title_usage.columns and 'total_item_requests' in ebsco_title_usage.columns:
        top_titles = ebsco_title_usage.groupby('title').agg({
            'total_item_requests': 'sum'
        }).reset_index().sort_values('total_item_requests', ascending=False).head(10)
        print("Top 10 Titles by Requests:")
        print(top_titles)
        print("\n" + "="*80 + "\n")
except Exception as e:
    print(f"Could not create top resources summaries: {e}")

# 3. Combined usage metrics across all platforms (if possible)
try:
    # Create a list of all dataframes with usage metrics
    all_usage_dfs = []
    
    # Add relevant columns from each dataframe if they exist
    if 'reporting_period' in ebsco_db_usage.columns and 'searches_regular' in ebsco_db_usage.columns:
        db_usage_summary = ebsco_db_usage.groupby('reporting_period')['searches_regular'].sum().reset_index()
        db_usage_summary.columns = ['date', 'ebsco_database_searches']
        all_usage_dfs.append(db_usage_summary)
        
    if 'reporting_period' in ebsco_platform_usage.columns and 'searches_regular' in ebsco_platform_usage.columns:
        platform_usage_summary = ebsco_platform_usage.groupby('reporting_period')['searches_regular'].sum().reset_index()
        platform_usage_summary.columns = ['date', 'ebsco_platform_searches']
        all_usage_dfs.append(platform_usage_summary)
    
    # If we have data to combine
    if all_usage_dfs:
        # Start with the first dataframe
        all_usage = all_usage_dfs[0]
        
        # Merge with the rest
        for df in all_usage_dfs[1:]:
            all_usage = pd.merge(all_usage, df, on='date', how='outer')
            
        # Sort by date
        all_usage = all_usage.sort_values('date')
        
        print("Combined Usage Metrics Across All Platforms:")
        print(all_usage)
except Exception as e:
    print(f"Could not create combined usage metrics: {e}")
```

```python
# Load all CSV files from data_processed folder
import pandas as pd
import os
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from datetime import datetime

# Set plotting style
plt.style.use('ggplot')
sns.set(font_scale=1.2)

# Dictionary to store all dataframes
processed_data = {}

# Load all CSV files from the data_processed folder
csv_files = [f for f in os.listdir('data_processed') if f.endswith('.csv')]
print(f"Loading {len(csv_files)} CSV files from data_processed folder...")

for file in csv_files:
    name = os.path.splitext(file)[0]  # Remove .csv extension
    file_path = os.path.join('data_processed', file)
    processed_data[name] = pd.read_csv(file_path)
    print(f"Loaded {name}: {processed_data[name].shape[0]} rows, {processed_data[name].shape[1]} columns")

# List of all loaded dataframes
print("\nLoaded dataframes:")
for name, df in processed_data.items():
    print(f"- {name}")
```

```python
# Basic analysis of each dataframe
for name, df in processed_data.items():
    print(f"\n{'='*80}\n{name.upper()} ANALYSIS\n{'='*80}")
    
    # Basic statistics
    print("\nBasic Statistics:")
    print(f"- Shape: {df.shape}")
    print(f"- Columns: {', '.join(df.columns.tolist())}")
    
    # Check for missing values
    missing = df.isnull().sum()
    if missing.sum() > 0:
        print("\nMissing Values:")
        print(missing[missing > 0])
    else:
        print("\nNo missing values found.")
    
    # Data types
    print("\nData Types:")
    print(df.dtypes)
    
    # Summary statistics for numeric columns
    numeric_cols = df.select_dtypes(include=['number']).columns
    if len(numeric_cols) > 0:
        print("\nNumeric Column Statistics:")
        print(df[numeric_cols].describe().T)
    
    # Sample data
    print("\nSample Data:")
    print(df.head(3))
    
    # Check for date columns to identify time series data
    date_cols = [col for col in df.columns if 'date' in col.lower() or 'period' in col.lower() or 'month' in col.lower() or 'year' in col.lower()]
    if date_cols:
        print(f"\nPotential date columns found: {date_cols}")
        
    print("\n" + "-"*80)
```

```python
# Calculate Year-over-Year (YoY) percentage changes for visitor data
import pandas as pd
import numpy as np

# Check if commons_usage data is available
if 'commons_usage' in processed_data:
    df = processed_data['commons_usage'].copy()
    print("Calculating visitor percentage changes year-over-year...")
    
    # Check for date column
    date_cols = [col for col in df.columns if 'date' in col.lower()]
    if date_cols:
        date_col = date_cols[0]
        
        # Convert to datetime if needed
        if not pd.api.types.is_datetime64_any_dtype(df[date_col]):
            try:
                df[date_col] = pd.to_datetime(df[date_col])
                print(f"Converted {date_col} to datetime format")
            except Exception as e:
                print(f"Error converting date: {e}")
        
        if pd.api.types.is_datetime64_any_dtype(df[date_col]):
            # Extract year and month for grouping
            df['year'] = df[date_col].dt.year
            df['month'] = df[date_col].dt.month
            df['month_name'] = df[date_col].dt.month_name()
            df['year_month'] = df[date_col].dt.strftime('%Y-%m')
            
            # Find columns that might represent visitor counts
            visitor_cols = [col for col in df.columns if any(term in col.lower() for term in ['visitor', 'count', 'traffic', 'in', 'out', 'ins', 'outs'])]
            
            if not visitor_cols:
                # If no specific visitor columns found, use numeric columns
                visitor_cols = df.select_dtypes(include=['number']).columns.tolist()
                # Remove year and month columns from the list
                visitor_cols = [col for col in visitor_cols if col not in ['year', 'month']]
            
            if visitor_cols:
                print(f"Found potential visitor count columns: {visitor_cols}")
                
                # Calculate monthly totals
                monthly_totals = df.groupby(['year', 'month', 'month_name'])[visitor_cols].sum().reset_index()
                print("\nMonthly visitor totals:")
                print(monthly_totals.head())
                
                # Calculate year-over-year percentage changes
                yoy_changes = []
                
                # For each year and month
                for year in monthly_totals['year'].unique():
                    for month in range(1, 13):  # 1-12 for Jan-Dec
                        # Current period data
                        current = monthly_totals[(monthly_totals['year'] == year) & (monthly_totals['month'] == month)]
                        
                        # Previous year same month
                        previous = monthly_totals[(monthly_totals['year'] == year - 1) & (monthly_totals['month'] == month)]
                        
                        if not current.empty and not previous.empty:
                            for col in visitor_cols:
                                current_val = current[col].values[0]
                                previous_val = previous[col].values[0]
                                
                                # Calculate percentage change
                                if previous_val != 0:
                                    pct_change = ((current_val - previous_val) / previous_val) * 100
                                else:
                                    pct_change = np.nan
                                
                                month_name = current['month_name'].values[0]
                                
                                yoy_changes.append({
                                    'year': year,
                                    'month': month,
                                    'month_name': month_name,
                                    'metric': col,
                                    'current_value': current_val,
                                    'previous_value': previous_val,
                                    'pct_change': pct_change
                                })
                
                # Convert to DataFrame
                if yoy_changes:
                    yoy_df = pd.DataFrame(yoy_changes)
                    print("\nYear-over-Year percentage changes:")
                    print(yoy_df.head())
                    
                    # Calculate overall year-to-date (YTD) change
                    # Group by year and calculate totals
                    yearly_totals = df.groupby('year')[visitor_cols].sum().reset_index()
                    
                    print("\nYearly visitor totals:")
                    print(yearly_totals)
                    
                    # Calculate YoY changes for each year
                    yearly_changes = []
                    for year in sorted(yearly_totals['year'].unique()):
                        if year > min(yearly_totals['year']):
                            current_year_data = yearly_totals[yearly_totals['year'] == year]
                            prev_year_data = yearly_totals[yearly_totals['year'] == year - 1]
                            
                            if not current_year_data.empty and not prev_year_data.empty:
                                for col in visitor_cols:
                                    current_val = current_year_data[col].values[0]
                                    prev_val = prev_year_data[col].values[0]
                                    
                                    # Calculate percentage change
                                    if prev_val != 0:
                                        pct_change = ((current_val - prev_val) / prev_val) * 100
                                    else:
                                        pct_change = np.nan
                                    
                                    yearly_changes.append({
                                        'year': year,
                                        'metric': col,
                                        'current_value': current_val,
                                        'previous_value': prev_val,
                                        'pct_change': pct_change
                                    })
                    
                    if yearly_changes:
                        yearly_changes_df = pd.DataFrame(yearly_changes)
                        print("\nYearly percentage changes:")
                        print(yearly_changes_df)
                        
                        # Save the results for use in the dashboard
                        yoy_df.to_csv('data_processed/monthly_yoy_changes.csv', index=False)
                        yearly_changes_df.to_csv('data_processed/yearly_yoy_changes.csv', index=False)
                        
                        print("\nYoY change data saved to data_processed folder")
                        
                        # Display the most recent year's change for the dashboard image
                        latest_year = yearly_changes_df['year'].max()
                        latest_year_change = yearly_changes_df[yearly_changes_df['year'] == latest_year]
                        
                        if not latest_year_change.empty:
                            for _, row in latest_year_change.iterrows():
                                print(f"\nFor dashboard image - {row['year']} Visitors:")
                                print(f"Total: {int(row['current_value']):,}")
                                print(f"YoY Change: {row['pct_change']:.1f}%")
                else:
                    print("No year-over-year data available for comparison")
            else:
                print("No visitor count columns found in the data")
        else:
            print("Could not convert date column to datetime format")
    else:
        print("No date column found in commons_usage data")
else:
    print("Commons usage data not found. Please ensure the data is loaded correctly.")
```

```python
# Calculate additional growth metrics for the dashboard
print("Calculating additional growth metrics...")

# Dictionary to store all growth metrics
growth_metrics = {}

# 1. Resource Utilization Growth
if 'ebsco_db_usage' in processed_data and 'ebsco_platform_usage' in processed_data:
    try:
        # Combine database and platform usage if possible
        db_df = processed_data['ebsco_db_usage']
        platform_df = processed_data['ebsco_platform_usage']
        
        # Check for date/period columns
        date_cols_db = [col for col in db_df.columns if 'date' in col.lower() or 'period' in col.lower()]
        date_cols_platform = [col for col in platform_df.columns if 'date' in col.lower() or 'period' in col.lower()]
        
        if date_cols_db and date_cols_platform:
            date_col_db = date_cols_db[0]
            date_col_platform = date_cols_platform[0]
            
            # Convert to datetime if needed
            if not pd.api.types.is_datetime64_any_dtype(db_df[date_col_db]):
                db_df[date_col_db] = pd.to_datetime(db_df[date_col_db])
            
            if not pd.api.types.is_datetime64_any_dtype(platform_df[date_col_platform]):
                platform_df[date_col_platform] = pd.to_datetime(platform_df[date_col_platform])
            
            # Extract year and month
            db_df['year'] = db_df[date_col_db].dt.year
            db_df['month'] = db_df[date_col_db].dt.month
            platform_df['year'] = platform_df[date_col_platform].dt.year
            platform_df['month'] = platform_df[date_col_platform].dt.month
            
            # Find usage columns
            usage_cols_db = [col for col in db_df.columns if 'searches' in col.lower() or 'requests' in col.lower()]
            usage_cols_platform = [col for col in platform_df.columns if 'searches' in col.lower() or 'requests' in col.lower()]
            
            if usage_cols_db and usage_cols_platform:
                # Calculate resource utilization: searches per database
                db_counts = db_df.groupby('year')['database'].nunique().reset_index()
                db_counts.columns = ['year', 'database_count']
                
                # Calculate total searches per year
                db_searches = db_df.groupby('year')[usage_cols_db].sum().reset_index()
                
                # Merge the counts and searches
                db_utilization = pd.merge(db_counts, db_searches, on='year')
                
                # Calculate searches per database
                for col in usage_cols_db:
                    db_utilization[f'searches_per_db_{col}'] = db_utilization[col] / db_utilization['database_count']
                
                # Calculate year-over-year growth in searches per database
                for year in sorted(db_utilization['year'].unique())[1:]:  # Skip the first year
                    current_year = db_utilization[db_utilization['year'] == year]
                    prev_year = db_utilization[db_utilization['year'] == year - 1]
                    
                    if not current_year.empty and not prev_year.empty:
                        for col in usage_cols_db:
                            metric_name = f'searches_per_db_{col}'
                            current_val = current_year[metric_name].values[0]
                            prev_val = prev_year[metric_name].values[0]
                            
                            if prev_val != 0:
                                pct_change = ((current_val - prev_val) / prev_val) * 100
                                
                                # Store in growth metrics
                                if 'resource_utilization' not in growth_metrics:
                                    growth_metrics['resource_utilization'] = []
                                
                                growth_metrics['resource_utilization'].append({
                                    'year': year,
                                    'metric': f'Searches per Database ({col})',
                                    'current_value': current_val,
                                    'previous_value': prev_val,
                                    'pct_change': pct_change
                                })
                
                print("Resource utilization growth metrics calculated")
    except Exception as e:
        print(f"Error calculating resource utilization growth: {e}")

# 2. User Engagement Growth (based on title requests)
if 'ebsco_title_usage' in processed_data:
    try:
        title_df = processed_data['ebsco_title_usage']
        
        # Check for date/period columns
        date_cols = [col for col in title_df.columns if 'date' in col.lower() or 'period' in col.lower()]
        
        if date_cols:
            date_col = date_cols[0]
            
            # Convert to datetime if needed
            if not pd.api.types.is_datetime64_any_dtype(title_df[date_col]):
                title_df[date_col] = pd.to_datetime(title_df[date_col])
            
            # Extract year and month
            title_df['year'] = title_df[date_col].dt.year
            title_df['month'] = title_df[date_col].dt.month
            
            # Find request columns
            request_cols = [col for col in title_df.columns if 'request' in col.lower()]
            
            if request_cols:
                # Calculate unique titles accessed per year
                title_counts = title_df.groupby('year')['title'].nunique().reset_index()
                title_counts.columns = ['year', 'unique_titles']
                
                # Calculate total requests per year
                title_requests = title_df.groupby('year')[request_cols].sum().reset_index()
                
                # Merge the counts and requests
                title_engagement = pd.merge(title_counts, title_requests, on='year')
                
                # Calculate average requests per title
                for col in request_cols:
                    title_engagement[f'requests_per_title_{col}'] = title_engagement[col] / title_engagement['unique_titles']
                
                # Calculate year-over-year growth in requests per title
                for year in sorted(title_engagement['year'].unique())[1:]:  # Skip the first year
                    current_year = title_engagement[title_engagement['year'] == year]
                    prev_year = title_engagement[title_engagement['year'] == year - 1]
                    
                    if not current_year.empty and not prev_year.empty:
                        for col in request_cols:
                            metric_name = f'requests_per_title_{col}'
                            current_val = current_year[metric_name].values[0]
                            prev_val = prev_year[metric_name].values[0]
                            
                            if prev_val != 0:
                                pct_change = ((current_val - prev_val) / prev_val) * 100
                                
                                # Store in growth metrics
                                if 'user_engagement' not in growth_metrics:
                                    growth_metrics['user_engagement'] = []
                                
                                growth_metrics['user_engagement'].append({
                                    'year': year,
                                    'metric': f'Requests per Title ({col})',
                                    'current_value': current_val,
                                    'previous_value': prev_val,
                                    'pct_change': pct_change
                                })
                
                print("User engagement growth metrics calculated")
    except Exception as e:
        print(f"Error calculating user engagement growth: {e}")

# 3. Digital vs. Physical Usage Ratio Growth
if 'commons_usage' in processed_data and 'ebsco_platform_usage' in processed_data:
    try:
        commons_df = processed_data['commons_usage']
        platform_df = processed_data['ebsco_platform_usage']
        
        # Check for date columns
        commons_date_cols = [col for col in commons_df.columns if 'date' in col.lower()]
        platform_date_cols = [col for col in platform_df.columns if 'date' in col.lower() or 'period' in col.lower()]
        
        if commons_date_cols and platform_date_cols:
            commons_date_col = commons_date_cols[0]
            platform_date_col = platform_date_cols[0]
            
            # Convert to datetime if needed
            if not pd.api.types.is_datetime64_any_dtype(commons_df[commons_date_col]):
                commons_df[commons_date_col] = pd.to_datetime(commons_df[commons_date_col])
            
            if not pd.api.types.is_datetime64_any_dtype(platform_df[platform_date_col]):
                platform_df[platform_date_col] = pd.to_datetime(platform_df[platform_date_col])
            
            # Extract year
            commons_df['year'] = commons_df[commons_date_col].dt.year
            platform_df['year'] = platform_df[platform_date_col].dt.year
            
            # Find visitor and usage columns
            visitor_cols = [col for col in commons_df.columns if any(term in col.lower() for term in ['visitor', 'count', 'traffic', 'in', 'out', 'ins', 'outs'])]
            usage_cols = [col for col in platform_df.columns if 'searches' in col.lower() or 'requests' in col.lower()]
            
            if not visitor_cols:
                # If no specific visitor columns found, use numeric columns
                visitor_cols = commons_df.select_dtypes(include=['number']).columns.tolist()
                visitor_cols = [col for col in visitor_cols if col not in ['year', 'month']]
            
            if visitor_cols and usage_cols:
                # Calculate yearly totals
                commons_yearly = commons_df.groupby('year')[visitor_cols].sum().reset_index()
                platform_yearly = platform_df.groupby('year')[usage_cols].sum().reset_index()
                
                # Find common years
                common_years = set(commons_yearly['year']).intersection(set(platform_yearly['year']))
                
                # Calculate digital to physical ratio for common years
                digital_physical_ratio = []
                
                for year in common_years:
                    commons_data = commons_yearly[commons_yearly['year'] == year]
                    platform_data = platform_yearly[platform_yearly['year'] == year]
                    
                    for v_col in visitor_cols:
                        for u_col in usage_cols:
                            physical_usage = commons_data[v_col].values[0]
                            digital_usage = platform_data[u_col].values[0]
                            
                            # Calculate ratio of digital to physical
                            if physical_usage != 0:
                                ratio = digital_usage / physical_usage
                                
                                digital_physical_ratio.append({
                                    'year': year,
                                    'physical_metric': v_col,
                                    'digital_metric': u_col,
                                    'physical_usage': physical_usage,
                                    'digital_usage': digital_usage,
                                    'digital_to_physical_ratio': ratio
                                })
                
                # Convert to DataFrame
                if digital_physical_ratio:
                    ratio_df = pd.DataFrame(digital_physical_ratio)
                    
                    # Calculate year-over-year growth in ratio
                    for v_col in visitor_cols[:1]:  # Use first visitor column
                        for u_col in usage_cols[:1]:  # Use first usage column
                            filtered_ratio = ratio_df[(ratio_df['physical_metric'] == v_col) & 
                                                    (ratio_df['digital_metric'] == u_col)]
                            
                            # Sort by year
                            filtered_ratio = filtered_ratio.sort_values('year')
                            
                            # Calculate growth
                            for i in range(1, len(filtered_ratio)):
                                current_year = filtered_ratio.iloc[i]['year']
                                current_ratio = filtered_ratio.iloc[i]['digital_to_physical_ratio']
                                prev_ratio = filtered_ratio.iloc[i-1]['digital_to_physical_ratio']
                                
                                if prev_ratio != 0:
                                    pct_change = ((current_ratio - prev_ratio) / prev_ratio) * 100
                                    
                                    # Store in growth metrics
                                    if 'digital_physical_ratio' not in growth_metrics:
                                        growth_metrics['digital_physical_ratio'] = []
                                    
                                    growth_metrics['digital_physical_ratio'].append({
                                        'year': current_year,
                                        'metric': f'Digital to Physical Usage Ratio',
                                        'current_value': current_ratio,
                                        'previous_value': prev_ratio,
                                        'pct_change': pct_change
                                    })
                
                print("Digital vs. Physical usage ratio growth metrics calculated")
    except Exception as e:
        print(f"Error calculating digital vs. physical ratio growth: {e}")

# Save all growth metrics to CSV files
for metric_type, data in growth_metrics.items():
    if data:
        df = pd.DataFrame(data)
        file_path = f'data_processed/{metric_type}_growth.csv'
        df.to_csv(file_path, index=False)
        print(f"Saved {metric_type} growth metrics to {file_path}")

# Create a summary of all growth metrics
all_growth_metrics = []
for metric_type, data in growth_metrics.items():
    for item in data:
        item['category'] = metric_type
        all_growth_metrics.append(item)

if all_growth_metrics:
    all_growth_df = pd.DataFrame(all_growth_metrics)
    all_growth_df.to_csv('data_processed/all_growth_metrics.csv', index=False)
    print("Saved combined growth metrics to data_processed/all_growth_metrics.csv")
    
    # Display the latest growth metrics for each category
    latest_year = all_growth_df['year'].max()
    latest_growth = all_growth_df[all_growth_df['year'] == latest_year]
    
    print(f"\nLatest Growth Metrics ({latest_year}):")
    for category in latest_growth['category'].unique():
        print(f"\n{category.replace('_', ' ').title()}:")
        category_metrics = latest_growth[latest_growth['category'] == category]
        for _, row in category_metrics.iterrows():
            print(f"  - {row['metric']}: {row['current_value']:.2f} ({row['pct_change']:.1f}% vs. previous year)")
else:
    print("No growth metrics were calculated")
```
